```python
SPGC2024 = SPGC2024.rename(columns={
    'rank': 'vi_tri_xep_hang',
    'previous_rank': 'vi_tri_tuan_truoc',
    'peak_rank': 'vi_tri_cao_nhat',
    'streams': 'luot_nghe',
    'weeks_on_chart': 'so_tuan_tren_bang_xep_hang',
    'artist_names': 'nghe_si',
    'track_name': 'ten_bai_hat',
    'uri': 'spotify_uri',
    'Source.Name': 'nguon_du_lieu_1',
    'is_new_entry': 'bai_hat_moi',
    'rank_change': 'thay_doi_thu_hang'
})
```

```python
SPGC2024.head()
```

```python
nguon_csv = (
    SPGC2024[['nguon_du_lieu_1']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
)

nguon_csv.head()
```

```python
hang_record = (
    SPGC2024[['source']]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
)

hang_record.head()
```

```python
SPGC2024_main = SPGC2024.drop(columns=['nguon_du_lieu_1', 'source','spotify_uri'])
```

```python
nguon_csv.to_csv(
    '../data/processed/bang_nguon_csv.csv',
    index=False
)

hang_record.to_csv(
    '../data/processed/bang_hang_record.csv',
    index=False
)
```

Chuẩn hoá text (artist, song)

```python
SPGC2024_main['nghe_si'] = (
    SPGC2024_main['nghe_si']
    .str.strip()
)

SPGC2024_main['ten_bai_hat'] = (
    SPGC2024_main['ten_bai_hat']
    .str.strip()
)
```

Loại bỏ outliers không hợp lý

```python
SPGC2024_main['luot_nghe'].describe()
```

Nhóm độ phổ biến theo lượt nghe (bins)

```python
SPGC2024_main['nhom_luot_nghe'] = pd.qcut(
    SPGC2024_main['luot_nghe'],
    q=4,
    labels=['Thấp', 'Trung bình', 'Cao', 'Rất cao']
)
```

Phân loại độ bền bài hát (longevity)

```python
def phan_loai_do_ben(weeks):
    if weeks <= 4:
        return 'Ngắn hạn'
    elif weeks <= 12:
        return 'Trung hạn'
    else:
        return 'Dài hạn'

SPGC2024_main['do_ben'] = SPGC2024_main['so_tuan_tren_bang_xep_hang'].apply(phan_loai_do_ben)
```
